```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
hidrograma = pd.read_excel('/content/hodrograma1.xlsx')
hidrograma.head()
```

```python
missing_values_count = hidrograma.isnull().sum()
```

```python
missing_values_count[0:10]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.lineplot(data=hidrograma, x='PENDIENTE', y='Q', palette='Set1')
plt.title('Pendiente vs Caudal', fontsize=16)
plt.xlabel('Pendiente', fontsize=14)
plt.ylabel('Caudal', fontsize=14)
plt.gcf().set_facecolor('#008080')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.histplot(data=hidrograma, x='PENDIENTE', y='Q', palette='Set1')
plt.title('Pendiente vs Caudal', fontsize=16)
plt.xlabel('Pendiente', fontsize=14)
plt.ylabel('Caudal', fontsize=14)
plt.gcf().set_facecolor('#008080')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data=hidrograma, x='PENDIENTE', y='Q', palette='Set1')
plt.title('Pendiente vs Caudal', fontsize=16)
plt.xlabel('pendiente', fontsize=14)
plt.ylabel('caudal', fontsize=14)
plt.gcf().set_facecolor('#008080')
plt.show()
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.boxenplot(data=hidrograma, x='PENDIENTE', y='Q', palette='Set1')
plt.title('Pendiente vs caudal', fontsize=16)
plt.xlabel('pendiente', fontsize=14)
plt.ylabel('caudal', fontsize=14)
plt.gcf().set_facecolor('#008080')
plt.show()
```
