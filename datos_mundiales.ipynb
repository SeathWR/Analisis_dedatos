```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
year = [1930, 1934, 1938, 1950, 1954, 1958, 1962, 1966, 1970, 1974, 1978, 1982, 1986, 1990, 1994, 1998, 2002, 2006, 2010, 2014, 2018]
```

```python
def get_partidos(year):
  web = f"https://en.wikipedia.org/wiki/{year}_FIFA_World_Cup"
respuesta = requests.get(web)
content = respuesta.text
soup = BeautifulSoup(content, "lxml")
partidos = soup.find_all("div", class_="footballbox")


for match in partidos:
    home.append(match.find("th", class_="fhome").get_text())
    score.append(match.find("th", class_="fscore").get_text())
    away.append(match.find("th", class_="faway").get_text())

dict_football = {'home': home, 'score': score, 'away': away}
df_football = pd.DataFrame(dict_football)
df_football['year'] = year
return df_football
```
