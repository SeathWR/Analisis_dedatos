```python
flight_filepath = "/content/flight_delays.csv"
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

```python
flight_data.head()
```

#GRAFICO DE BARRAS

```python
plt.figure(figsize=(10,6))
plt.title("Vuelos atrasados por mes")
#Grfico de barras que muestra el retraso promedio en la llegada de los vuelos de Spirit Airlines por mes
sns.barplot(x=flight_data.index, y=flight_data['NK'])

plt.ylabel("retraso en minutos")
```

#HEATMAP

```python
plt.figure(figsize=(14,7))
plt.title("Vuelos atrasados por mes")
#Grfico de mapa de calor que muestra el retraso promedio en la llegada de los vuelos de Spirit Airlines por mes
sns.heatmap(data=flight_data,annot=True)
plt.xlabel("aerolinea")
```

```python
iris_filepath = "/content/iris.csv"
iris_data = pd.read_csv(iris_filepath)
```

```python
iris_data.head()
```

#histogramas

```python
sns.histplot(data=iris_data['Petal Length (cm)'])
```

```python
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
```

```python
#2d kde plots
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
#
```

```python
#histogramas por especies
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')
plt.title('Histograma de Longitud de Ptalos por Especie')
```

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)
plt.title('Kernel Density Estimation de Longitud de Ptalos por Especie')
#
```
