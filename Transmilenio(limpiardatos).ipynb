```python
import pandas as pd

df = pd.read_excel('/content/ValidacionesAgo2024.xlsx')
```

```python
df.head()
```

```python
df['Estación'] = df['Estación'].str.replace(r"\(.*?\)\sC\d+\|", '', regex=True)
df['Línea'] = df['Línea'].str.replace(r"\(.*?\)\sC\d+\|", '', regex=True)
df['Acceso de Estación'] = df['Acceso de Estación'].str.replace(r"\(.*?\)\sC\d+\|", '', regex=True)
```

```python
df['Estación'] = df['Estación'].str.replace(r"\(.*?\)", '', regex=True)
df['Línea'] = df['Línea'].str.replace(r"\(.*?\)", '', regex=True)
df['Acceso de Estación'] = df['Acceso de Estación'].str.replace(r"\(.*?\)", '', regex=True)
```

```python
df.head()
```

```python
df.to_excel('ValidacionesAgo2024F.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10,16))
sns.countplot(y= 'Estación', data=df, order=df['Estación'].value_counts(15).index)
```

```python
plt.figure(figsize=(10,16))
plt.title('Estaciones con mas validaciones')
sns.heatmap(data=df['Estación'].value_counts().head(15).to_frame(),annot=True)
```

```python
plt.figure(figsize=(10,16))
plt.title('Estaciones con mas validaciones')
sns.heatmap(data=df['Acceso de Estación'].value_counts().head(15).to_frame(),annot=True)
```
